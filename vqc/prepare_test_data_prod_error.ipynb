```python
path = "./data/telesale/v2/[VQC] Dataset cho AI engine.xlsx"
data_df = pd.read_excel(path, engine="openpyxl", sheet_name="NLP - TLS - 2")
data_df = data_df[["Sample", "Actual intent\n(old model)", "Expectation intent"]]
data_df = data_df[1:]
data_df
```

```python
X = data_df["Sample"].tolist()

with open("./data/telesale/v2/nlp_tls_2.txt", "w") as f:
    for text in X:
        f.write(text)
        f.write("\n")
```

```python
y_true = data_df["Expectation intent"].tolist()

try:
    with open("./data/telesale/v2/nlp_tls_2_label.txt", "w") as f:
        for label in y_true:
            f.write(label)
            f.write("\n")
except:
    print(label)
```
