```python
from sklearn import datasets
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap

from utils import tsne
```

```python
digits = datasets.load_digits()
```

```python
# Try visualizing an image
image = digits.images[7]
plt.figure(figsize=(3, 3))
plt.matshow(image, fignum=0)
plt.show()
```

```python
tsne = TSNE(n_components=2)
embedded_digits = tsne.fit_transform(digits.data)
```

```python
embedded_digits.shape
```

```python
embedded_digits[0]
```

```python
x = embedded_digits[:, 0]
y = embedded_digits[:, 1]

# colors mapping
colors_dict = {
    0: "#5ef3da",
    1: "#5ec4f3",
    2: "#5e72f3",
    3: "#925ef3",
    4: "#c95ef3",
    5: "#f35e74",
    6: "#f38d5e",
    7: "#f3c45e",
    8: "#d3f35e",
    9: "#557740"
}
colors = ListedColormap(list(colors_dict.values()))

# Plot scatter
fig, ax = plt.subplots(figsize=(15, 10))
scatter = plt.scatter(x, y,
                      s=13, # maker size
                      c=digits.target, # maker color
                      cmap=colors) # color mapping

# Add legend
legend = ax.legend(*scatter.legend_elements(),
                   title="Digits")
ax.add_artist(legend)

# Final show
plt.show()
```
